```python
#pandas is a library used for working data sets, it has functions for analysing, manuplating and exploring the data. tables, files 
import pandas as pd
```

```python
#data loaded into dataset - on pd class - read csv function call to pass it to dataset
dataset=pd.read_csv("Salary_data.csv")
```

```python
#excuted the dataset variable and verify the same out put as per excel data.
dataset
```

```python
#input variable - we are spliting input variable value - YearsExperiance
independent=dataset[["YearsExperience"]]
```

```python
independent
```

```python
#output variable - we are spliting output variable value - salary data
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
#here trainning and test set splitting  -using sklearn library we are splitting training and test set.valures return it
#giving input and output variables to split and getting output.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent, dependent,test_size=0.30,random_state=0)
```

```python
#model creation - it contains weight and bias formulas
#regressor is a variable   regressor.fit - function - training set input/output sent this is learning phase
from sklearn.linear_model import LinearRegression 
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# weight calculation - slope
weight=regressor.coef_
weight
```

```python
bais=regressor.intercept_
bais
```

```python
#End goal - predection 
y_pred=regressor.predict(x_test)
```

```python
#to find the r2 value
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#To save the model using pickle and creating a dummy file name
import pickle 
filename="finalized_model_linear.sav"
```

```python
#dumping the model on the file, wb - write binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
# and save the model and load and check it for sending phase two for deployment.  rb - read binar
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[13]])
```
